TEST 2

Since these wind farms are offshore, it can take maintenance engineers up to a day to travel out to a turbine and get it back online after a fault.
If we could detect signs of this fault mode emerging before the turbine shuts down, we could send an engineer out ahead of time, minimising downtime.
Experts have suggested that an abrupt increase in vibration in the turbine (measured by sensors RadialVibX and RadialVibY) might provide a good predictor.
Write a script that will generate a time series chart that illustrates an example of this.

```python
dr1 = '/home/flavia/Test_personali/data-test/time_series/'

data_turbine = pd.read_csv(dr1 + "3057.csv")
data_turbine
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean_X = np.mean(data_turbine['RadialVibX'], axis=0)
mean_Y = np.mean(data_turbine['RadialVibY'], axis=0)
SD_X = np.std(data_turbine['RadialVibX'], axis=0)
SD_Y = np.std(data_turbine['RadialVibY'], axis=0)

# Create histograms for both columns
plt.hist(data_turbine['RadialVibX'], bins=20, density=True, alpha=0.5, color='blue', label='VibX')
plt.hist(data_turbine['RadialVibY'], bins=20, density=True, alpha=0.5, color='orange', label='VibY')

# Plot vertical lines for means and mean ± standard deviations
plt.axvline(mean_X, color='blue', linestyle='dashed', linewidth=2, label='Mean X')
plt.axvline(mean_X + SD_X, color='blue', linestyle='dashed', linewidth=2, label='SDX_1')
plt.axvline(mean_X - SD_X, color='blue', linestyle='dashed', linewidth=2, label='SDX_2')

plt.axvline(mean_Y, color='orange', linestyle='dashed', linewidth=2, label='Mean Y')
plt.axvline(mean_Y + SD_Y, color='orange', linestyle='dashed', linewidth=2, label='SDY_1')
plt.axvline(mean_Y - SD_Y, color='orange', linestyle='dashed', linewidth=2, label='SDY_2')

# Add labels, title, and legend
plt.xlabel('VibX')
plt.ylabel('VibY')
plt.title('Test')
plt.legend()

# Show the plot
plt.show()
```
